```python
import pandas as pd


# Load dataset
df = pd.read_csv('marketing_campaign.csv', sep='\t')  # Tab-separated file

# Basic inspection
print("Shape:", df.shape)
df.head()
```

```python
# Data types and null values
df.info()

# Summary statistics
df.describe()
```

```python
# Check for null values
df.isnull().sum()

# Remove rows with missing values
df = df.dropna()

# Check for duplicates
df.duplicated().sum()

# Drop duplicates
df = df.drop_duplicates()
```

```python
# Check unique values for categorical fields
print(df['Education'].value_counts())
print(df['Marital_Status'].value_counts())

# Standardize categories (if needed)
df['Marital_Status'] = df['Marital_Status'].replace({
    'Alone': 'Single',
    'Absurd': 'Single',
    'YOLO': 'Single'
})
```

```python
# Outlier detection using IQR


Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['Income'] >= Q1 - 1.5 * IQR) & (df['Income'] <= Q3 + 1.5 * IQR)]
```

```python
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram: Age
current_year = datetime.now().year
df['Age'] = current_year - df['Year_Birth']

plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Spending habits boxplot
spend_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts']
df[spend_cols].plot(kind='box', figsize=(10, 6), title="Boxplot of Spending on Products")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot: Income vs Spending
sns.pairplot(df, vars=['Income', 'MntWines', 'MntMeatProducts'], hue='Education')
plt.show()
```
